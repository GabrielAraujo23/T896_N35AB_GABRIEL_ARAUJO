Importar as bibliotecas que serão utilizadas no Projeto

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

Carregar a imagem original do Projeto para iniciar o pré processamento.

```python
imagem_original = cv2.imread('img_folha_7.JPG')
plt.imshow(imagem_original)
plt.axis('off')
```

Converter do Padrão BGR para o RGB

```python
imagem_original_rgb = cv2.cvtColor(imagem_original, cv2.COLOR_BGR2RGB)
imagem_mediana = cv2.medianBlur(imagem_original_rgb, 3)
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.imshow(imagem_mediana)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(imagem_original_rgb)
plt.axis('off')
```

Decompor a imagem de RGB para HSV

```python
imagem_decomposta = cv2.cvtColor(imagem_mediana, cv2.COLOR_RGB2HSV)
plt.imshow(imagem_decomposta)
plt.axis('off')
```

Utilizar máscaras verdes e escuras


```python
# H, S, V = cv2.split(imagem_decomposta)

lower_green = np.array([35, 40, 40])
upper_green = np.array([85, 255, 255])
mascara_verde = cv2.inRange(imagem_decomposta, lower_green, upper_green)
parte_verde = cv2.bitwise_and(imagem_mediana, imagem_mediana, mask=mascara_verde)


lower_dark = np.array([0, 10, 0])
upper_dark = np.array([40, 255, 130])
mascara_escura = cv2.inRange(imagem_decomposta, lower_dark, upper_dark)
parte_escura = cv2.bitwise_and(imagem_mediana, imagem_mediana, mask=mascara_escura)
```

Separar os subplots do H S e V 

```python
plt.figure(figsize=(20,8))
plt.subplot(1, 2, 1)
plt.imshow(parte_verde, cmap = 'gray')
plt.title('Parte Verde')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(parte_escura, cmap = 'gray')
plt.title('Parte Escura')
plt.axis('off')
```
